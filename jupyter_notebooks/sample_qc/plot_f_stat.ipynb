```python
#plot histogram of f_stat from sex imputation

import hail as hl
import pyspark
from hail.plot import show
import bokeh
from bokeh.plotting import output_file, save
import os

sc = pyspark.SparkContext()
tmp_dir = "hdfs://spark-master:9820/"
lustre_dir = "file:///lustre/scratch123/qc/BiB/"
hl.init(sc=sc, tmp_dir=lustre_dir, default_reference="GRCh38")

hl.plot.output_notebook()

tmp_file = "/lustre/scratch123/qc/BiB/plots/sample_qc/tmp.html"#temp file for plots
```

```python
sex_mt_file = lustre_dir + "matrixtables/mt_sex_annotated.mt"
sex_mt = hl.read_matrix_table(sex_mt_file)
qc_ht = sex_mt.cols()
qc_ht.show()
```

```python
p = hl.plot.histogram(qc_ht.f_stat, legend="f_stat", title="ALSPAC sample QC: f_stat")
show(p)

plot_file = "/lustre/scratch123/qc/BiB/plots/sample_qc/f_stat.html"
output_file(tmp_file)
save(p)
os.rename(tmp_file, plot_file)
```
