```python
import hail as hl
import pyspark
from hail.plot import show
hl.plot.output_notebook()

sc = pyspark.SparkContext()
tmp_dir = "hdfs://spark-master:9820/"
lustre_dir = "file:///lustre/scratch123/qc/"
hl.init(sc=sc, tmp_dir=tmp_dir, default_reference="GRCh38")
```

```python
mtdir = lustre_dir + "matrixtables/"
mtfile = mtdir + "/non_stratified_sample_qc_dp20_gq20_vaf0_25.mt"
mt = hl.read_matrix_table(mtfile)
mt.count()
```

```python
mt.entries().show()
```

```python
#filter to indels only
mtindel = mt.filter_rows(hl.is_indel(mt.alleles[0], mt.alleles[1]))
mtindel.count()
```

```python
#annotate with indel length
mtindel = mtindel.annotate_rows(indel_length=(hl.len(mtindel.alleles[1]) - hl.len(mtindel.alleles[0])))
mtindel.entries().show()
```

```python
#filter to Sanger and broad, and remove hom ref
sangermt = mtindel.filter_entries(mtindel.sequencing_location == 'Sanger')
sangermt = sangermt.filter_entries(sangermt.GT.is_hom_ref(), keep = False)
broadmt = mtindel.filter_entries(mtindel.sequencing_location == 'Bristol')
broadmt = broadmt.filter_entries(broadmt.GT.is_hom_ref(), keep = False)
```

```python
p = hl.plot.histogram(sangermt.entries().indel_length, title='Indel length histogram Sanger samples', bins = 40, range=(-20,20), legend='Indel length')
show(p)
```

```python
p = hl.plot.histogram(broadmt.entries().indel_length, title='Indel length histogram Broad samples', bins = 40, range=(-20,20), legend='Indel length')
show(p)
```

```python
#look at common/rare indels per person - annotate with MAF from gnomad (split multi first)
sangermt = hl.split_multi_hts(sangermt)
```

```python
broadmt = hl.split_multi_hts(broadmt)
```

```python
#load gnomad data
resourcedir = lustre_dir + "resources/"
gnomadfile = resourcedir + "gnomad_v3-0_AF.ht"
gnomad_ht = hl.read_table(gnomadfile)
gnomad_ht.show()
```

```python
#annotate with gnomad
sangermt = sangermt.annotate_rows(gnomad_maf=gnomad_ht[sangermt.row_key].maf)
```

```python
broadmt = broadmt.annotate_rows(gnomad_maf=gnomad_ht[broadmt.row_key].maf)
```

```python
sangermt.entries().show()
```

```python
#filter to common and rare
sangermt_common = sangermt.filter_rows(sangermt.gnomad_maf > 0.05)
sangermt_common.count()
```

```python
sangermt_common.entries().show()
```

```python
sangermt_rare = sangermt.filter_rows(sangermt.gnomad_maf <= 0.05)
```

```python
broadmt_common = broadmt.filter_rows(broadmt.gnomad_maf > 0.05)
broadmt_rare = broadmt.filter_rows(broadmt.gnomad_maf <= 0.05)
```

```python
#run sample QC on broad and sanger common and rare indel sets. Hopefully this will over-write previous sample QC columns
sangermt_common = hl.sample_qc(sangermt_common)
sangermt_common.entries().show()
```

```python
sangermt_rare = hl.sample_qc(sangermt_rare)
```

```python
broadmt_common = hl.sample_qc(broadmt_common)
broadmt_rare = hl.sample_qc(broadmt_rare)
broadmt_common.cols().show()
```

```python
#remove unwanted columns
broad_common_f = broadmt_common.filter_cols(broadmt_common.sequencing_location == "Bristol")
broad_common_f.cols().show()
```

```python
broadmt_common = broadmt_common.filter_cols(broadmt_common.sequencing_location == "Bristol")
broadmt_rare = broadmt_rare.filter_cols(broadmt_rare.sequencing_location == "Bristol")
sangermt_common = sangermt_common.filter_cols(sangermt_common.sequencing_location == "Sanger")
sangermt_rare = sangermt_rare.filter_cols(sangermt_rare.sequencing_location == "Sanger")
```

```python
p = hl.plot.histogram(sangermt_common.sample_qc.n_deletion, title='Sanger samples: Number of common deletions per sample')
show(p)
```

```python
p = hl.plot.histogram(sangermt_rare.sample_qc.n_deletion, title='Sanger samples: Number of rare deletions per sample')
show(p)
```

```python
p = hl.plot.histogram(broadmt_common.sample_qc.n_deletion, title='Broad samples: Number of common deletions per sample')
show(p)
```

```python
p = hl.plot.histogram(broadmt_rare.sample_qc.n_deletion, title='Broad samples: Number of rare deletions per sample')
show(p)
```
