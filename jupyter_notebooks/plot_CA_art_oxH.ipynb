```python
import hail as hl
import pyspark
from hail.plot import show
hl.plot.output_notebook()

from bokeh.plotting import output_file, save, reset_output

sc = pyspark.SparkContext()
tmp_dir = "hdfs://spark-master:9820/"
lustre_dir = "file:///lustre/scratch123/qc/"
hl.init(sc=sc, tmp_dir=tmp_dir, default_reference="GRCh38")
```

```python
statsfile = lustre_dir + "compare_art_oxH_fract_CA/art_oxH_and_ca_per_sample.txt"
ht=hl.import_table(statsfile, types={'art_oxh':'float','fraction_ca':'float'})
ht.show()
```

```python
p = hl.plot.scatter(ht.art_oxh,
                    ht.fraction_ca,
                    label=ht.run,
                    title='art_oxH vs fraction CA by run id', xlabel='art_oxH', ylabel='fraction CA', collect_all = True)
show(p)
```
