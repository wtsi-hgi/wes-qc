```python
import hail as hl
import pyspark
from hail.plot import show
hl.plot.output_notebook()

from bokeh.plotting import output_file, save, reset_output

sc = pyspark.SparkContext()
tmp_dir = "hdfs://spark-master:9820/"
lustre_dir = "file:///lustre/scratch123/qc/"
hl.init(sc=sc, tmp_dir=tmp_dir, default_reference="GRCh38")
```

```python
datafile = "file:///lustre/scratch123/qc/plots/precision_recall.txt"
ht = hl.import_table(datafile, impute = True)
ht.show()
```

```python
p = hl.plot.scatter(ht.snv_recall, ht.snv_precision, xlabel='Recall', ylabel='Precision', title="SNV precision/recall", label=ht.bin)
```

```python
show(p)
```
