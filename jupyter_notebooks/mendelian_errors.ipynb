```python
#look for mendelian errors in our data
import hail as hl
import pyspark

sc = pyspark.SparkContext()
tmp_dir = "hdfs://spark-master:9820/"
lustre_dir = "file:///lustre/scratch123/qc/"
hl.init(sc=sc, tmp_dir=tmp_dir, default_reference="GRCh38")
```

```python
pedfile = lustre_dir + "resources/trios.ped"
ped = hl.Pedigree.read(pedfile)
```

```python
mtfile = lustre_dir + "matrixtables/mt_varqc_splitmulti.mt"
mt = hl.read_matrix_table(mtfile)
mt.entries().show()
```

```python
#mendelian errors
all_errors, per_fam, per_sample, per_variant = hl.mendel_errors(mt['GT'], ped)
```

```python
#plot a histogram of the errors per family
from bokeh.io import show, output_notebook
output_notebook()

p = hl.plot.histogram(per_fam.errors, legend = 'Mendelian error count', title='Mendelian errors per family')
show(p)
```

```python
p = hl.plot.histogram(per_fam.snp_errors)
show(p)
```
